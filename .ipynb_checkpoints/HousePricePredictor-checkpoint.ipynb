# **House Price Predictor - A Machine Learning Project**

**Prosjektbeskrivelse:** 

I dette prosjektet har jeg prøvd å lære meg litt maskinlæring. Jeg har utviklet tre forskjellige maskinlæringsmodeller. Hver av dem representerer en distinkt tilnærming til problemet med å forutsi huspriser basert på ulike egenskaper ved eiendommene. Ved å bruke et datasett med informasjon om husets pris, arealet av huset, antall bad, antall soverom, størrelse på tomten, året huset ble bygd og om huset har peis eller ikke, utforsker jeg 3 maskinlæringsalgoritmer for å identifisere den mest nøyaktige metoden for å predikere huspriser. Prosjektet inkluderer datainnhenting, datarensing, modelltrening, evaluering av ytelse, og visualiseringer. Modellene kan og kunne blitt forbedret ytteligere men dette prosjektet er ment mest for min egen læring.

# **Load Data**
Datasettet er hentet fra kaggle.com: https://www.kaggle.com/datasets/shubhendra7/houseprices-dataset

```python
import pandas as pd

df = pd.read_csv('houseprices.csv')

df # a overview of the data in the dataset
```

# **Data Preperation**

## Check for null values in the dataset
Sjekker for nullverdier fordi mange maskinlæringsalgoritmer ikke kan håndtere nullverdier direkte. Hvis det er nullverdier i datasettet kan dette føre til at algoritmen krasjer eller gir svært unøyaktige resultater. 

```python
df.isnull().sum()
```

*Siden ingen av kolonnene i datasettet inneholder nullverdier kan vi fortsette uten å måtte håndtere nullverdier*

## Data Seperation into X and y
Deler datasetettet etter hva modellen skal lære (X-variablene) og hva den skal forutsi/predikere (y). X representerer alle kolonnene i datasettet (utenom pris) som kan påvirke prisen, som f.eks. antall soverom, bad, tomtstørrelse, osv. Modellen bruker disse uavhengige variablene til å lære mønstre som den bruker til å forutsi y (pris).

```python
y = df['Price'] 
X = df.drop('Price', axis=1) 
```

## Data Splitting
Når man trener en modell, er det viktig at modellen ikke bare lærer å huske dataene den ser. Hvis du trener og tester modellen på det samme datasettet, kan modellen lære spesifikke mønstre i det datasettet (overfitting) og ikke generalisere godt til usette data. Ved å dele opp dataene, sikrer man at modellens ytelse blir målt på data den ikke har sett før.

Derfor deler jeg datasettet inn i trenings- og testsett slik at jeg kan vurdere hvor godt modellen vil prestere på nye, usette data. Treningssettet brukes til å tilpasse modellen, mens testsettet hjelper med å vurdere modellens reelle ytelse.

I virkelige applikasjoner vil en modell bli brukt på nye data (data som modellen ikke har sett før). *Testsettet etterligner dette scenariet* ved å gi et estimat på hvordan modellen ville oppført seg i produksjon på usette data.

20% av dataene brukes som testsett, mens 80% brukes som treningssett. `Random_state=100` sikrer at hvis man kjører notatboken på nytt, vil trenings-testdelingen alltid være den samme, noe som gir konsistente resultater.

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=100)
```

## Normalize the features of the dataset
Ulike egenskaper/variabler kan ha forskjellige enheter og skalaer. For eksempel kan Living Area være målt i kvadratmeter, mens Age er målt i år. Hvis en variabel har en mye større skala enn en annen, kan det påvirke hvordan modeller lærer fra dataene. Normalisering bringer alle variablene inn i et likt skaleringsområde. Uten normalisering kan høyverdige variabler dominere over lavverdige funksjoner i modellen.
Dette kan føre til at noen modeller sliter med å finne den optimale løsningen fordi de vil tillegge noen variabler mer betydning enn andre

Noen modeller, som nevrale nettverk, er avhengige av at dataene er normalisert for å fungere optimalt. Uten normalisering kan aktiveringsfunksjonene (som ReLU) oppføre seg uventet.

Her bruker jeg en `StandardScaler` som transformerer variablene slik at de har en gjennomsnittlig verdi på 0 og en standardavvik på 1. 

```python
from sklearn.preprocessing import StandardScaler 

scaler = StandardScaler()  # Can also use MinMaxScaler()

# Fit the scaler on the training data and transform both training and test sets
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

# **Model Building**
Implementerer ulike modeller for å løse problemet med å forutsi huspriser

## Linear Regression

### Training the model
Den lineære regresjonsmodellen (lr) skal trenes  med treningsdataene fra datasettet, slik at den kan lære å identifisere sammenhenger mellom egenskapene/variablene (f.eks. størrelse, antall rom) og husprisene.

```python
from sklearn.linear_model import LinearRegression

# Create a Linear Regression model
lr = LinearRegression()

# Fit the model to the training data
lr.fit(X_train_scaled, y_train)
```

### Applying the model to make a prediction
Her brukes den trente lineære regresjonsmodellen (lr) til å lage prediksjoner for treningssettet (`X_train_scaled`). Altså modellen tar informasjonen fra treningssettet og gir en pris for hvert hus basert på det den har lært om sammenhengen mellom variablene.

På samme måte som for treningssettet, brukes den trente modellen til å lage prediksjoner for testsettet (`X_test_scaled`). Modellen benytter variablene i testsettet for å beregne de predikerte husprisene. 

```python
y_lr_train_pred = lr.predict(X_train_scaled) 
y_lr_test_pred = lr.predict(X_test_scaled)
```

### Evaluate model preformance
Bruker MSE og R² fordi de passer bra for evaluering av regressjonsoppgaver

**MSE** beregner den gjennomsnittlige kvadrerte feilen mellom de faktiske husprisene og de predikerte husprisene. Dette gjøres for både treningssettet og testsettet for å evaluere modellens ytelse.
- For **treningssettet** (`y_train` og `y_lr_train_pred`) gir MSE et mål på hvor godt modellen har lært fra treningsdataene. Jo lavere verdi, desto bedre er modellen til å forutsi husprisene basert på treningsdataene.
- For **testsettet** (`y_test` og `y_lr_test_pred`) gir MSE et mål på hvordan modellen generaliserer til nye, usette data. Jo lavere MSE for testsettet, desto bedre er modellen til å generalisere og forutsi huspriser på nye data

**R²** måler andelen av variansen i den avhengige variabelen (huspriser) som kan forklares av de uavhengige variablene (alle andre egenskapene ved huset) i modellen.
- Gir en prosentandel som er lett å tolke: for eksempel, en R² på 0.65 betyr at 65% av variasjonen i huspriser kan forklares av modellen.
- Høyere R²-verdi indikerer en bedre tilpasning til dataene, og hjelper til med å vurdere modellens generelle ytelse.

```python
from sklearn.metrics import mean_squared_error, r2_score

# For treningssettene
lr_train_mae = mean_squared_error(y_train, y_lr_train_pred) 
lr_train_r2 = r2_score(y_train, y_lr_train_pred)

# For testsettene
lr_test_mae = mean_squared_error(y_test, y_lr_test_pred) 
lr_test_r2 = r2_score(y_test, y_lr_test_pred)
```

**The results shown in a dataframe:**


```python
lr_results = pd.DataFrame(['Linear Regression', lr_train_mae, lr_train_r2, lr_test_mae, lr_test_r2]).transpose()
lr_results.columns = ['Method', 'Training MSE', 'Training R2', 'Test MSE', 'Test R2']

lr_results
```

**Training Mean Squared Error (MSE):** I dette tilfellet er MSE relativt høy, noe som antyder at det er en betydelig forskjell mellom de predikerte husprisene og de faktiske prisene i treningsdataene.

**Training R² Score:** En R² på 0,65 antyder at omtrent 65 % av variasjonen i husprisene kan forklares av modellen. Dette er et moderat nivå av forklaring, som indikerer at selv om modellen fanger opp noen relasjoner, kan det fortsatt være faktorer som påvirker husprisene som ikke blir tatt i betraktning.

**Test Mean Squared Error (MSE):** Test MSE er høyere enn trenings MSE, noe som kan indikere overfitting. Dette betyr at modellen presterer bedre på treningsdataene enn på testdataene, noe som tyder på at den ikke generaliserer godt til nye, usette data

**Test Mean Squared Error (MSE):** Test R² på 0.6246 er lavere enn trenings R², noe som bekrefter at modellen er mindre effektiv på testsettet. Dette indikerer at omtrent 62% av variasjonen i testdataene kan forklares av modellen, noe som er en nedgang fra treningssettet.

### Visualization of Predictions

```python
import matplotlib.pyplot as plt
import numpy as np

# Plotting Actual vs Predicted values
plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_lr_test_pred, color='blue', alpha=0.5, label='Predicted Prices')
plt.plot(y_test, y_test, color='red', linewidth=2, label='Actual Prices') 
plt.title('Actual vs Predicted House Prices (Linear Regression)')
plt.xlabel('Actual Prices')
plt.ylabel('Predicted Prices')
plt.legend()
plt.grid()
plt.show()
```

### Plotting the residuals. 
Residualer er forskjellen mellom de faktiske verdiene (observasjonene) og de predikerte verdiene fra en statistisk modell. De representerer hvor mye modellen feiler i sine prediksjoner

Hver blå prikk representerer en prediksjon: residual = faktisk pris − predikert pris:
- Hvis en residual er over den røde linjen (som representerer en residual på 0), har modellen undervurdert prisen.
- Hvis en residual er under den røde linjen, har modellen overvurdert prisen.

```python
# Calculate residuals
residuals_lr = y_test - y_lr_test_pred

# Plotting the residuals:
plt.figure(figsize=(10, 6))
plt.scatter(y_lr_test_pred, residuals_lr, color='blue', alpha=0.5)
plt.axhline(0, color='red', linestyle='--')
plt.title('Residuals of Linear Regression')
plt.xlabel('Predicted Prices')
plt.ylabel('Residuals')
plt.grid()
plt.show()
```

**Tolkning av visualiseringen**:
There seems to be a slight fan-shaped pattern, where residuals are more spread out as the predicted price increases. This indicates that your linear regression model might have trouble predicting house prices as they increase, meaning the model could be underfitting for higher-priced homes.
No obvious curve: There doesn't appear to be a strong curvature pattern, so the relationship between features and target values seems to be mostly linear.
Some outliers: There are a few extreme points where the model either underestimates or overestimates by large amounts. These outliers might need further investigation or could represent unusual properties.

## Decision Tree Regression

### Training the model

```python
from sklearn.tree import DecisionTreeRegressor

# Create a Decision Tree Regressor model
dt = DecisionTreeRegressor(random_state=100)

# Fit the model to the training data
dt.fit(X_train_scaled, y_train)
```

### Visualization of feature importance 
Denne visualiseringen hjelper med å forstå hvilke features som bidrar mest med oppgaven om å forutsi pris på hus

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Get feature importance
feature_importances = dt.feature_importances_
features = X.columns

# Plotting the feature importance
plt.figure(figsize=(8, 6))
sns.barplot(x=feature_importances, y=features)
plt.title('Feature Importance')
plt.xlabel('Importance')
plt.ylabel('Features')
plt.show()
```

*Større søyler indikerer funksjoner som var mer innflytelsesrike i å bestemme husprisene i Decision Tree-modellen.*

### Applying the model to make a prediction

```python
y_dt_train_pred = dt.predict(X_train_scaled) 
y_dt_test_pred = dt.predict(X_test_scaled)
```

### Evaluate model performance

```python
# For training sets
dt_train_mae = mean_squared_error(y_train, y_dt_train_pred)  # Mean Squared Error
dt_train_r2 = r2_score(y_train, y_dt_train_pred)  # R² score

# For test sets
dt_test_mae = mean_squared_error(y_test, y_dt_test_pred)  # Mean Squared Error
dt_test_r2 = r2_score(y_test, y_dt_test_pred)  # R² score
```

**The results shown in a dataframe:**

```python
dt_results = pd.DataFrame(['Decision Tree', lr_train_mae, lr_train_r2, lr_test_mae, lr_test_r2]).transpose()
dt_results.columns = ['Method', 'Training MSE', 'Training R2', 'Test MSE', 'Test R2']

dt_results
```

### Plotting the residuals. 
Residualer er forskjellen mellom de faktiske verdiene (observasjonene) og de predikerte verdiene fra en statistisk modell. De representerer hvor mye modellen feiler i sine prediksjoner

Hver blå prikk representerer en prediksjon: residual = faktisk pris − predikert pris:

Hvis en residual er over den røde linjen (som representerer en residual på 0), har modellen undervurdert prisen.
Hvis en residual er under den røde linjen, har modellen overvurdert prisen.

```python
# Calculate residuals
residuals_dt = y_test - y_dt_test_pred

# Plotting the residuals:
plt.figure(figsize=(10, 6))
plt.scatter(y_dt_test_pred, residuals_dt, color='blue', alpha=0.5)
plt.axhline(0, color='red', linestyle='--')
plt.title('Residuals of Decision Tree')
plt.xlabel('Predicted Prices')
plt.ylabel('Residuals')
plt.grid()
plt.show()
```

**Tolkning av visualiseringen:**

Grafen viser  en viss spredning rundt null-linjen, men det ser ut til å være klumper, spesielt for lavere huspriser (under 150,000). Dette antyder at modellen kan ha problemer med å håndtere bestemte prisområder og kanskje ikke klarer å generalisere like godt i hele datasettet.

Noen punkter (outliers) ligger langt unna null-linjen, noe som indikerer at modellen enten sterkt under- eller overestimerer huspriser i disse tilfellene. Disse outliers kan potensielt være unøyaktige data eller tilfeller som ikke passer godt inn i mønstrene beslutningstreet har lært.

## Neural network model and Deep Learning
Huspriser påvirkes av ulike faktorer (egenskaper) som kan ha ikke-lineære sammenhenger med prisen. Nevrale nettverk kan fange opp komplekse ikke-lineære mønstre i data gjennom flere lag av sammenkoblede nevroner.

Modellen er kompilert med Adam optimizer and mean squared error as the loss function som er standard for regresjonsproblemer.

Det er mulig å justere modellen og eksperimentere med ulike antall lag, nevroner, læringsrater og regulariseringsparametere for å finne den beste konfigurasjonen. Jeg har sagt meg fornøyd med slik modellen er nå

### Create, Compile and Train the model

```python
# Import necessary libraries
from keras.models import Sequential
from keras.layers import Dense, Dropout, Input
from keras.regularizers import l2 # This can help in making the model generalize better on test data.

# Create the Neural Network model
nn_model = Sequential()

# Input layer based on the number of features in the scaled training data
nn_model.add(Input(shape=(X_train_scaled.shape[1],)))  
# Hidden layers with L2 regularization and adjusted neurons
nn_model.add(Dense(512, activation='relu', kernel_regularizer=l2(0.01)))  
nn_model.add(Dropout(0.3)) # Add dropout layer to prevent overfitting
nn_model.add(Dense(256, activation='relu', kernel_regularizer=l2(0.01)))
nn_model.add(Dropout(0.3)) # Add dropout layer to prevent overfitting
nn_model.add(Dense(128, activation='relu', kernel_regularizer=l2(0.01)))
nn_model.add(Dropout(0.3)) # Add dropout layer to prevent overfitting

# Output layer for regression (1 neuron)
nn_model.add(Dense(1))
 
# Compile the model
nn_model.compile(optimizer='adam', loss='mean_squared_error', metrics=['mean_absolute_error'])

# Train the model
history = nn_model.fit(X_train_scaled, y_train, epochs=400, batch_size=16, validation_split=0.2)
```

### Applying the model to make a prediction

```python
# Predictions
y_nn_train_pred = nn_model.predict(X_train_scaled)
y_nn_test_pred = nn_model.predict(X_test_scaled)
```

### Evaluate the model

```python
# Evaluate model performance
nn_train_mse = mean_squared_error(y_train, y_nn_train_pred)  # Mean Squared Error
nn_train_r2 = r2_score(y_train, y_nn_train_pred)  # R² score

nn_test_mse = mean_squared_error(y_test, y_nn_test_pred)  # Mean Squared Error
nn_test_r2 = r2_score(y_test, y_nn_test_pred)  # R² score

# Create a DataFrame to display results
nn_results = pd.DataFrame(['Neural Network', nn_train_mse, nn_train_r2, nn_test_mse, nn_test_r2]).transpose()
nn_results.columns = ['Method', 'Training MSE', 'Training R2', 'Test MSE', 'Test R2']

# Print the results
print(nn_results)
```

### Visualization of Predictions for Neural Network
Jo nærmere punktene er den diagonale linjen, desto bedre er modellens prediksjoner.

```python
plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_nn_test_pred, color='purple', alpha=0.5, label='Predicted Prices')
plt.plot(y_test, y_test, color='red', linewidth=2, label='Actual Prices')  # Diagonal line for reference
plt.title('Actual vs Predicted House Prices (Neural Network)')
plt.xlabel('Actual Prices')
plt.ylabel('Predicted Prices')
plt.legend()
plt.grid()
plt.show()
```
